```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clean = pd.read_csv('Cleaned.csv', index_col=0)
```

```python
clean.head()
```

```python
clean.columns
```

Education

```python
sns.countplot(x='Education', data=clean)
```

Marital Status

```python
sns.countplot(x='Marital_Status', data=clean)
```

Place

```python
each_place = ['NumWebPurchases','NumCatalogPurchases', 'NumStorePurchases']
total_place = []
[total_place.append(len(clean[clean[place] == 1])) for place in each_place]

plt.figure(figsize=(15,5))
sns.barplot(x=each_place, y=total_place)
plt.title('Total purchases of each place')
```

```python
each_prods = ['MntWines','MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds']
total_prods = []
[total_prods.append(len(clean[clean[prod] == 1])) for prod in each_prods]

plt.figure(figsize=(15,5))
sns.barplot(x=each_prods, y=total_prods)
plt.title('Total amount of each products')
```

```python
each_cmp = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
total_cmp = []
[total_cmp.append(len(clean[clean[cmp] == 1])) for cmp in each_cmp]

plt.figure(figsize=(15,5))
sns.barplot(x=each_cmp, y=total_cmp)
plt.title('Total of Accepted Campaign offers on each Campaigns')
```

```python
sns.countplot(x='Complain', data=clean)
```

```python
colors = sns.color_palette('pastel')[0:5]
Generation = clean['Generation'].value_counts()
plt.figure(figsize=(10,5))
plt.pie(Generation, labels = Generation.index, colors = colors, autopct='%1.2f%%')
plt.title('Generation')
```

```python
clean_dist = clean[['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 
       'Age', 'Amount_Total', 'Total_Purchases', 'Children', 'Family_Size', 'Total_AcceptedCmp', 'Days_Active']]

for i in clean_dist.columns:
    sns.displot(x=i, bins=30, data=clean)
```

```python
sns.heatmap(data=clean_dist.corr(), cmap='coolwarm')
```
